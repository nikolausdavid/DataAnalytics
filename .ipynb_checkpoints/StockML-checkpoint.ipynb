### Data Analytics using Python and Machine Learning Models

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
```

```python
dax = yf.Ticker("^GDAXI").history(start="2000-01-01")
```

```python
dax.head() 
```

```python
del dax["Dividends"]
del dax["Stock Splits"]
```

```python
plt.plot(dax["Close"]);
```

```python
dax["Tomorrow"] = dax["Close"].shift(-1)
```

```python
dax["Target"] = (dax["Tomorrow"] > dax["Close"]).astype(int)
dax
```

```python
model = RandomForestClassifier(n_estimators=150,
                              min_samples_split=100,
                              random_state=1)
train = dax.iloc[:-100]
test = dax.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
preds = pd.Series(model.predict(test[predictors]), index=test.index)
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
plt.plot(combined)
plt.tick_params(axis='x', rotation=60);
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = pd.Series(model.predict(test[predictors]), index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)                
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
        
    return pd.concat(all_predictions)
```

```python
predictions = backtest(dax, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
horizons = [2,5,60,250, 1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = dax.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    dax[ratio_column] = dax["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    dax[trend_column] = dax.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors += [ratio_column, trend_column]
```

```python
dax = dax.dropna(subset=dax.columns[dax.columns != "Tomorrow"])
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds,index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)                
    return combined
```

```python
predictions = backtest(dax, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```
