```python
## Modules
"https://pypi.org/project/yahooquery/1.1.0/"
import yfinance as yf 
import json
from yahooquery import Ticker  
import datetime
```

```python
## Functions

def print_titles(x, y=8):
    list = []
    for i in range(y):
        list.append(x[i]["title"])
    return list
        
```

```python
tsla = yf.Ticker("TSLA")
```

```python
news = tsla.news
```

```python
#tsla.incomestmt
```

```python
[print(i) for i in print_titles(news)]
```
